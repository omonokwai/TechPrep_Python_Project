```python
# importing pandas library
import pandas as pd
```

```python
#import data set
df = pd.read_csv('Grades_Short.csv')
```

```python
# View the First 5 rows of the Data set
df.head()
```

```python
# Explore the data 
df.describe()
```

```python
# Create a new column
df['Avg_Score']= df.drop(['ID','Name'], axis =1).mean(axis=1)
```

```python
# View the First 5 rows of the Data set
df.head()
```

```python
# Create the function for assigning grades
def grades(Avg_score):
    if Avg_score > 90:
        return 'A+'
    elif Avg_score >80:
        return 'A'
    elif Avg_score >70:
        return 'B'
    elif Avg_score > 60:
        return 'C'
    elif Avg_score > 55:
        return 'D'
    else:
        return 'F'
```

```python
# Create The Final Grade Column
df['Final_Grade'] = df['Avg_Score'].apply(grades)
```

```python
#Print the up Modified Data Set
print(df)
```

```python
#Save The New Data

df.to_csv('Grades_Short_Modified.csv')

print('The modified version as been saved as "Grades_Short_Modified.csv"')
```
